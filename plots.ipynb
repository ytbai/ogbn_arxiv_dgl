```
import torch
from torch import nn
import numpy as np
import matplotlib.pyplot as plt

from data_factory import *
from model_factory import *
```

```
# Load ogbn-arxiv dataset
# Store graph onto CPU device
# Save dataset in directory root
dataset_name = "ogbn-arxiv"
device = "cpu:0"
root = "dataset/"
data_factory = DataFactory(dataset_name, device, root)
```

```
# Make figure
n_bins = 40

train_labels = data_factory.labels[data_factory.get_idx_split("train")].flatten().numpy()
valid_labels = data_factory.labels[data_factory.get_idx_split("valid")].flatten().numpy()
test_labels = data_factory.labels[data_factory.get_idx_split("test")].flatten().numpy()

stacked_labels = [train_labels, valid_labels, test_labels]
label = ["train", "valid", "test"]

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10,5))
ax.hist(stacked_labels, n_bins, stacked=True, label=label)
ax.legend()
ax.set_title("Node Frequency by Target Class")
fig.savefig("plots/freq.png")
```
